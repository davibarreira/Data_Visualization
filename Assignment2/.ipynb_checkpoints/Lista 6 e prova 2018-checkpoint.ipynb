```python
# Tamanho original
kb = M.shape[0]*M.shape[1]
```

```python
U,d,Vt = np.linalg.svd(M)
```

```python
S = np.zeros(M.shape)
np.fill_diagonal(S,d)
```

```python
(U@S@Vt - M).round(3)
```
